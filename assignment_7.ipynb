**What is the expected value of a single die roll?**

Let X represent the outcome of the first die. 

E[X] = (1 + 2 + 3 + 4 + 5 + 6) / 6 = 21 / 6 = 3.5

The expected value of a single die roll is 3.5.


**What is the expected value of rolling two dice and adding the results together?**

Let X represent the outcome of the first die. Let Y represent the outcome of the second die.

E[X] = E[Y] = (1 + 2 + 3 + 4 + 5 + 6) / 6 = 21 / 6 = 3.5

E[X + Y] = E[X] + E[Y] = 3.5 + 3.5 = 7

The expected value of the sum of two dice is 7. 

**What is the expected winnings of any gamble in European roulette?**

The European roulette has 37 slots so there are the number 1-36, plus a single 0. If you bet on a single number and it hits, the casino pays you 35 times your bet. If it doesn't hit, you lose your bet. 

Let X be the winnings on a $1 bet. 

There are two possible outcomes: 
- If you win, you get a $ \$35 $ profit. This outcome has a probability of 1/37. 
- If you lose, you lose the $ \$1 $ you bet. This outcome has a probability of 36/37. 

E[X] = (win amount * P(win)) + (loss amount * P(loss))

E[X] = (35 * (1/37) + (-1 * 36/37)) = (35/37) - (36/37) = (-1/37) = -0.027. 

On average, you would lose about 2.7 cents per $ \$1 $ bet in European roulette. This negative expected value applies to every type of bet on the table because the payouts are always slightly less than the true odds. The game is designed so the house always has a consistent edge, making every roulette gamble a losing bet in the long run.

Suppose $X$ has an expectation $\mathbb{E}[X]<\infty$ and variance $\mathbb{V}[X]<\infty$; this isn't always true, but is usually true
- Consider making a new variable, $\varepsilon = X - \mathbb{E}[X]$
- What's the expectation of $\varepsilon$?
    - E[$\varepsilon$] = E[X - E[X]] = E[X] - E[X] = 0
- What's the variance of $\varepsilon$?
    - V[$\varepsilon$] = V[X - E[X]]
    - Since E[X] is a constant, V[X - E[X]] = V[X] = $ \sigma_X^2$
- So we can write any random variable in the form $X = \mathbb{E}[X] + \varepsilon, $ where $\mathbb{E}[\varepsilon]=0$ and $\mathbb{V}[\varepsilon] = \sigma_X^2$. If that's true, show that we can also write any random variable in the form $X = \mathbb{E}[X] + \sigma_X \varepsilon$, where $\mathbb{E}[\varepsilon]=0$ and $\mathbb{V}[\varepsilon]=1$
    - Since $\sigma_X>0$, define $\varepsilon' = \dfrac{\varepsilon}{\sigma_X}$
    - the mean is E[$\varepsilon'$] = $\dfrac{1}{\sigma_X} E[\varepsilon]$ = 0
    - the variance is V[$\varepsilon'$] = $\dfrac{1}{\sigma_X^2} V[\varepsilon]$ = $\dfrac{\sigma_X^2}{\sigma_X^2} = 1$
    - So, $X = \mathbb{E}[X] + \varepsilon = \mathbb{E}[X] + \sigma_X \varepsilon'$
- Now replace $\mathbb{E}[X]$ with $x\beta$, and the stage is set for regression models
    - $X = x\beta + \sigma_X \varepsilon'$
    - $E[\varepsilon'] = 0$, $V[\varepsilon'] = 1$
